# Learning Outcomes

```python
# Read the tech_bootcamp sheet
bayesian_df = pd.read_excel(file_path, sheet_name='6040')

# Set the second row as the header
bayesian_df.columns = bayesian_df.iloc[2]

# Drop the first two rows
bayesian_df = bayesian_df.drop([0, 1, 2]).reset_index(drop=True)

# Display the cleaned DataFrame
print("Cleaned bayesian DataFrame:")
display(bayesian_df.head())
```

```python
# Read the tech_bootcamp sheet
tech_bootcamp_df = pd.read_excel(file_path, sheet_name='tech_bootcamp')

# Set the second row as the header
tech_bootcamp_df.columns = tech_bootcamp_df.iloc[0]

# Drop the first two rows
tech_bootcamp_df = tech_bootcamp_df.drop([0]).reset_index(drop=True)

# Display the cleaned DataFrame
print("Cleaned tech_bootcamp DataFrame:")
display(tech_bootcamp_df.head())
```

```python
# Define the course sheets
course_sheets = ['5001', '5012', '5100', '5110', '6001', '6002', '6003', '6011', '6012', '6013', '6021', '6030', '6050', 'biz_analytics']

# Function to clean a single sheet
def clean_dataframe(file_path, sheet_name):
    df = pd.read_excel(file_path, sheet_name=sheet_name)
    df.columns = df.iloc[1]
    df = df.drop([0, 1]).reset_index(drop=True)
    return df

# Read and clean the course sheets
dfs = {sheet: clean_dataframe(file_path, sheet) for sheet in course_sheets}

# Ensure that the columns are correctly named
for sheet_name, df in dfs.items():
    df.columns = df.columns.str.strip()
```

```python
# Initialize lists to store the mnemonics and learning outcomes
mnemonics = []
learning_outcomes = []

# Extract mnemonics and learning outcomes from each dataframe
for sheet_name, df in dfs.items():
    mnemonics.extend(df['MNEMONIC'])
    learning_outcomes.extend(df['LEARNING OUTCOME'])

mnemonics.extend(tech_bootcamp_df['MNEMONIC'])
learning_outcomes.extend(tech_bootcamp_df['LEARNING OUTCOME'])

mnemonics.extend(bayesian_df['MNEMONIC'])
learning_outcomes.extend(bayesian_df['LEARNING OUTCOME'])


# Create a new dataframe with the combined data
combined_df = pd.DataFrame({
    'MNEMONIC': mnemonics,
    'LEARNING OUTCOME': learning_outcomes
})

# Add the isactive column and set all values to True
combined_df['IsActive'] = True
```

```python
# Rename columns to match the Courses schema
combined_df = combined_df.rename(columns={
    'MNEMONIC': 'CourseMnemonic',
    'LEARNING OUTCOME': 'OutcomeDescription',
    'IsActive': 'IsActive',
})


# Reorder columns to match the Courses schema
combined_df = combined_df[['CourseMnemonic', 'OutcomeDescription', 'IsActive']]
```

# Instructors

```python
# Read the tech_bootcamp sheet
instructors_df = pd.read_excel(file_path, sheet_name='other_data_raw')
```

```python
# Set the second row as the header
instructors_df.columns = instructors_df.iloc[29]

# Drop the first two rows
instructors_df = instructors_df.drop(range(30)).reset_index(drop=True)
```

```python
# Function to split names
def split_names(names):
    if pd.isna(names):
        return []
    # Split by ',' or '&' and strip whitespace
    names = [name.strip() for name in names.replace('&', ',').split(',')]
    return names

all_names = [name for names in instructors_df.iloc[:, 0].apply(split_names) for name in names]

# Create a new DataFrame with each name in a separate row
cleaned_df = pd.DataFrame(all_names, columns=['InstructorName'])

# Remove duplicates
cleaned_df = cleaned_df.drop_duplicates()

# Add the 'IsActive' column based on the condition
cleaned_df['IsActive'] = ~cleaned_df['InstructorName'].isin(['Jeremy Bolton', 'Luis Felipe Rosado Murillo'])

# Display the cleaned DataFrame
print("Cleaned DataFrame:")
instructors_df_ready = cleaned_df
instructors_df_ready
```

# Assignments

```python
# Read the tech_bootcamp sheet
assignments_df = pd.read_excel(file_path, sheet_name='assigned_raw')

# Set the second row as the header
assignments_df.columns = assignments_df.iloc[1]

# Drop the first two rows
assignments_df = assignments_df.drop([0, 1]).reset_index(drop=True)
```

```python
# Extract Fall2021 DataFrame
fall2021_df = assignments_df.iloc[:, [0, 1, 2]]
fall2021_df['term'] = 'fall2021'
fall2021_df = fall2021_df.iloc[:-2].reset_index(drop=True)
fall2021_df = fall2021_df.iloc[:, [i for i in range(fall2021_df.shape[1]) if i != 1]]
fall2021_df = fall2021_df.rename(columns={fall2021_df.columns[0]: 'CourseMnemonic',
                                         fall2021_df.columns[1]: 'InstructorName', 
                                         fall2021_df.columns[2]: 'Term'})
```

# Instructor Assignments

```python
instructassign_df = assignments_df[['AssignmentID', 'InstructorName']]
```

```python
# Function to split instructors
def split_instructors(row):
    instructors = row['InstructorName']
    
    # Replace known typos and split instructors
    if ' & ' in instructors:
        instructors = instructors.replace(' & ', ', ')
    
    # Handle possible multi-instructor entries and typos
    instructors_list = [i.strip() for i in instructors.split(',') if i.strip()]
    
    # Return a DataFrame where each instructor is in a separate row
    return pd.DataFrame({
        'AssignmentID': [row['AssignmentID']] * len(instructors_list),
        'InstructorName': instructors_list
    })


# Apply the function and concatenate results
instructassign_df_ready = pd.concat(instructassign_df.apply(split_instructors, axis=1).tolist(), ignore_index=True)

instructassign_df_ready
```
